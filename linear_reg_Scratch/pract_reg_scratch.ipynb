```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
x_train = df['YearsExperience'].values
y_train = df['Salary'].values
```

```python
plt.scatter(x_train,y_train)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
def cost_function(x,y,w,b):
    m=len(x)
    cost_sum =0
    for i in range(m):
        f= w * x[i] + b
        cost = (f - y[i])**2
        cost_sum+=cost
    total_sum = (1/2*m)*cost_sum
    return total_sum
```

```python
def gradient_function(x,y,w,b):
    m=len(x)
    dc_dw=0
    dc_db=0
    for i in range(m):
        f= w * x[i] + b
        dc_dw += (f-y[i])*x[i]
        dc_db += f-y[i]

    dc_dw = dc_dw/m
    dc_db = dc_db/m
    return dc_dw,dc_db
```

```python
def gradient_descent(x,y,alpha,iterations):
    w=0
    b=0
    for i in range(iterations):
        dw,db =gradient_function(x,y,w,b)
        w = w- alpha * dw
        b = b-alpha*db
        print(f"Iteration {i}: Cost {cost_function(x, y, w, b)}\n")

    return w,b
```

```python
alpha=0.02
iterations=10000
final_w,final_b = gradient_descent(x_train,y_train,alpha,iterations)
print(f"w:{final_w}, b:{final_b}")
```

```python
plt.scatter(x_train,y_train,label='Data Points')
x_vals=np.linspace(min(x_train),max(x_train),100)
y_vals = final_w * x_vals + final_b
plt.plot(x_vals,y_vals,color='red',label='Linear Regression Line')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
