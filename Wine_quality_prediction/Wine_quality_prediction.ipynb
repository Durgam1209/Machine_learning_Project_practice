```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('winequalityN.csv')
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        df[col].fillna(df[col].mean(),inplace=True)
df.isnull().sum()
```

```python
df.replace({'white': 1, 'red': 0}, inplace=True)
```

```python

df.hist(bins=20,figsize=(10,10))
plt.show()
```

```python
plt.bar(df['quality'],df['alcohol'])
plt.xlabel('quality')
plt.ylabel('alcohol')
plt.show()
```

```python
# Convert 'object' columns to numerical if they represent numbers

for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col]=pd.to_numeric(df[col],errors='coerce')
        except:
            pass
plt.figure(figsize=(12,10))
sns.heatmap(df.corr()>0.7,
            annot=True,
            cbar=False)
plt.show()
```

```python
df=df.drop('total sulfur dioxide',axis=1)
```

```python
df['best quality']=[1 if x>5 else 0 for x in df.quality]
```

```python
df.head()
```

```python
features=df.drop(['quality','best quality'],axis=1)
features= features.fillna(features.mean())
target=df['best quality']

x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.2,random_state=40)


from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
x_train = imputer.fit_transform(x_train)
x_test = imputer.transform(x_test)

x_train.shape, x_test.shape
```

```python
norm  = MinMaxScaler()
x_train = norm.fit_transform(x_train)
x_test = norm.transform(x_test)
```

```python
models = [LogisticRegression(),SVC(kernel='rbf'),XGBClassifier()]

for i in range(3):
    models[i].fit(x_train,y_train)
    print(models[i],':')
    print('Training Accuracy: ',metrics.roc_auc_score(y_train,models[i].predict(x_train)))
    print('Validation Accuracy: ',metrics.roc_auc_score(y_test,models[i].predict(x_test)))
    print()
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

cm=confusion_matrix(y_test,models[2].predict(x_test))
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=models[2].classes_)
disp.plot()
plt.show()
```

```python
print(metrics.classification_report(y_test,
                                    models[2].predict(x_test)))
```
